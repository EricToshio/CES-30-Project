```python
import tensorflow as tf

import tensorflow.keras as keras

import numpy

import pandas

from datetime import datetime

from sklearn.model_selection import train_test_split

import scipy.stats
```

```python
%load_ext tensorboard.notebook
```

> Ou seja, só o "country" fica vazio de vez em quando (já tirei fora essa coluna)

```python
users = pandas.read_csv('train/user_data.csv', usecols=lambda x: x not in ['country', 'last_online_time_seconds', 'registration_time_seconds'])
users = pandas.concat([users, pandas.DataFrame({'user_id_num': list(range(len(users)))})], axis=1)
```

```python
users
```

```python
problems = pandas.read_csv('train/problem_data.csv')
problems = pandas.concat([problems, pandas.DataFrame({'problem_id_num': list(range(len(problems)))})], axis=1)
problems
```

> labels garantidamente todos preenchidos (já verifiquei. Comentado abaixo)

```python
labels = pandas.read_csv('train/train_submissions.csv')
# any(map(lambda value: any(list(map(lambda x: isinstance(x, float) and numpy.isnan(x), value))), labels.values))
labels
```

```python
aggregate = pandas.merge(pandas.merge(labels,users,on='user_id',how='inner'), problems, on='problem_id', how='inner')
aggregate
```

> Ótimo ! Todas os labels se referem a usuários+problemas existentes

```python
train_aggregate, test_validation_aggregate = train_test_split(aggregate, test_size=0.3, shuffle=True)
test_aggregate, validation_aggregate = train_test_split(test_validation_aggregate, test_size=0.5, shuffle=True)
len(train_aggregate), len(test_aggregate), len(validation_aggregate)
```

```python
max_attempts_range = 6
min_attempts_range = 1
user_embedding_dimensionality = 50
problem_embedding_dimensionality = 50
dropout_embedding_rate = 0.5
hidden_dimensionality = 10
dropout_hidden_rate = 0.5

user_input = keras.layers.Input(shape=(1, ), dtype='int32')
problem_input = keras.layers.Input(shape=(1, ), dtype='int32')

user_embedding = keras.layers.Embedding(len(users), user_embedding_dimensionality) # input length ?
problem_embedding = keras.layers.Embedding(len(problems), problem_embedding_dimensionality) # input length ?

concat = keras.layers.Concatenate()
reshape = keras.layers.Reshape((user_embedding_dimensionality + problem_embedding_dimensionality, ))

dropout_embedding = keras.layers.Dropout(dropout_embedding_rate)

hidden_dense = keras.layers.Dense(hidden_dimensionality, activation='relu', use_bias=True)

dropout_hidden = keras.layers.Dropout(dropout_hidden_rate)

sigmoid = keras.layers.Dense(1, activation='sigmoid', use_bias=True)
softmax = keras.layers.Softmax()
chooseone = keras.layers.Maximum()

output = keras.layers.Lambda(lambda x: x * (max_attempts_range - min_attempts_range) + min_attempts_range)

network = output(
     sigmoid(
         dropout_hidden(
            hidden_dense(
                dropout_embedding(
                    reshape(
                        concat([user_embedding(user_input), problem_embedding(problem_input)])
                    )
                )
            )
        )
    )
)

#network = softmax(
#    dropout_hidden(
#        hidden_dense(
#            dropout_embedding(
#                reshape(
#                    concat([user_embedding(user_input), problem_embedding(problem_input)])
#                )
#            )
#        )
#    )
#)

model = keras.models.Model(inputs=[user_input, problem_input], outputs=network)
model.compile('rmsprop', loss='mse', metrics=['mse'])
# model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
logdir="logs/fit/" + datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir)

model.fit(
    [
        numpy.ndarray.flatten(train_aggregate[['user_id_num']].values), 
        numpy.ndarray.flatten(train_aggregate[['problem_id_num']].values)
    ], 
    numpy.ndarray.flatten(train_aggregate[['attempts_range']].values) - 1, 
    validation_data=(
        (
            numpy.ndarray.flatten(test_aggregate[['user_id_num']].values),
            numpy.ndarray.flatten(test_aggregate[['problem_id_num']].values)
        ),
        numpy.ndarray.flatten(test_aggregate[['attempts_range']].values) - 1
    ),
    epochs=100, 
    batch_size=32,
    callbacks=[
        # keras.callbacks.EarlyStopping(monitor='val_loss', patience=4),
        tensorboard_callback
    ]
)
```

```python
validation_comparison = pandas.DataFrame([model.predict((
        numpy.ndarray.flatten(validation_aggregate[['user_id_num']].values), 
        numpy.ndarray.flatten(validation_aggregate[['problem_id_num']].values)
    )), numpy.ndarray.flatten(validation_aggregate[['attempts_range']].values)])
validation_comparison
```
